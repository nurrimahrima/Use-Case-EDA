# Missing Value Detection
Identifying any missing values in the data before perform any analysis. It's actually depend on the purpose of the performance, but if there is any missing values, we can cleanse and prepare it by not remove it from dataframe but doing imputation. If the missing values are numerical variable, we can impute it by its median, while if the missing values are categorical variable, we can impute it by its mode.


```python
#to see missing values
datavehicle.isnull().sum().sort_values(ascending=False)
```

We can see that the missing values are very large. 

```python
#to see percentage of missing values of each variables
percentmissing=round(datavehicle.isnull().sum()/len(datavehicle)*100,2).sort_values(ascending=False)
print(percentmissing)
```

If we check with its percentage, it can be seen that there are two columns that have missing values over 50%. With this condition, if the comparation of missing values columns with the available data is 60 : 40, we can drop the columns. 

```python
#remove the over-missing values variables
datavehicle=datavehicle.drop(['size','vin'],axis=1)
datavehicle.head()
```

```python
print(len(datavehicle.columns))
```

```python
datavehicle.isnull().sum().sort_values(ascending=False)
```

Because 'url' and 'image_url' columns are presented by link and it can't be analyze, we can drop it from dataframe

```python
datavehicle=datavehicle.drop(['url','image_url'],axis=1)
datavehicle.head()
```

Imputing Missing Values of Numerical Variable with Median of its columns, because if it imputes by its mean, it will tend to be outlier. 

```python
datavehicle.info(), datavehicle.isnull().sum().sort_values(ascending=False)
```

```python
median = datavehicle['odometer'].median()
datavehicle['odometer'].fillna(median, inplace=True)
median = datavehicle['weather'].median()
datavehicle['weather'].fillna(median, inplace=True)
datavehicle
```

Imputing Missing Values of Categorical Variable with Mode of its columns, because it is easier to find mode in categorical variable columns.

```python
datavehicle.info(), datavehicle.isnull().sum().sort_values(ascending=False)
```

```python
categorical=['condition', 'paint_color', 'cylinders', 'drive', 'manufacturer', 'make', 
             'state_code', 'county_name', 'fuel', 'transmission', 'title_status','state_fips','county_fips','year','type']

for mod in categorical:
    modus=datavehicle[mod].mode().values[0]
    datavehicle[mod]=datavehicle[mod].fillna(modus)
    
datavehicle
```

```python
#checking that all the data have no missing values anymore
datavehicle.isnull().sum().sort_values(ascending=False)
```

# Outlier Detection
Outlier detection can be seen by visualize univariate analysis with some plot. <br>
At first, we need to split the data 'datavehicle' into categorical data and numerical data.

```python
#define numerical variables
numerical= ['price','weather', 'odometer']
numerical = datavehicle[numerical]
numerical.head()
```

```python
#separate numerical variables from dataset, and the dataset will be categoric variable
categorical = datavehicle.drop(numerical, axis = 1)
categorical.head()
```

```python
numerical.describe()
```

```python
categorical.describe()
```

```python
#define the distributin plot of each of numerical variables to see outlier
#determine the size of the figure plot
fig=plt.figure(figsize=(13,12))
axes=330
#using boxplot to visualize each numerical variables in one single loop 
for num in numerical:
    axes+=1
    fig.add_subplot(axes)
    sns.boxplot(data=numerical, x=num)

plt.show()
```

From the result above it can be seen that there are some outlier in numerical data 'price' and 'odometer'.<br>
'price' and 'odometer' have positive skew of normal curve with distribution of the data tends to go right. For modelling purposes, a different approach to outliers could be to try creating a transformation of the data, either transform the value or transform the variable, such as creating a percentile version of the original field, transform the variable to create normal distribution.<br>
On this case (EDA), we will not make any transformation to data outlier.

# Correlation

```python
#create correlation with heatmap
corr = numerical.corr(method = 'pearson')
#convert correlation to numpy array
mask = np.array(corr)
#to mask the repetitive value for each pair
mask[np.tril_indices_from(mask)] = False
fig, ax = plt.subplots(figsize = (15,12))
fig.set_size_inches(20,10)
sns.heatmap(corr, 
            mask = mask, 
            vmax = 0.9, 
            square = True, 
            annot = True)
```

From the result above, it can be seen that correlation between variables is very small, almost equal to zero.So, it can be concluded that correlation between variables in 'datavehicle' is negative.
